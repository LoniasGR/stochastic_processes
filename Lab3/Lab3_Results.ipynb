## Παραδοτέο 1

```python
def create_tennis_chain(p_a):
    
    p_b = 1 - p_a
    
    init_probs_2  = {'0-0':1}

    P_table = {
        '0-0': {'15-0':p_a, '0-15':p_b},
        '15-0': {'30-0':p_a, '15-15':p_b},
        '0-15': {'15-15': p_a, '0-30':p_b},
        '30-0': {'40-0': p_a, '30-15': p_b},
        '15-15': {'30-15': p_a, '15-30': p_b},
        '0-30': {'15-30': p_a, '0-40': p_b},
        '40-0': {'GameA': p_a, '40-15': p_b},
        '30-15': {'40-15': p_a, 'Deuce': p_b},
        '15-30': {'Deuce': p_a, '15-40': p_b},
        '0-40': {'15-40': p_a, 'GameB': p_b},
        '40-15': {'GameA': p_a, 'AdvA': p_b},
        '15-40': {'AdvB': p_a, 'GameB': p_b},
        'Deuce': {'AdvA': p_a, 'AdvB': p_b},
        'GameA': {'GameA': 1},
        'AdvA': {'GameA': p_a, 'Deuce': p_b},
        'AdvB': {'Deuce': p_a, 'GameB': p_b},
        'GameB': {'GameB': 1}
    }

    mc2 = markov_chain(P_table, init_probs_2)
    return mc2
```

```python
x_axis = np.linspace(0.0, 1.0, 100)
y_axis = []

for p_a in x_axis:
    
    mc2 = create_tennis_chain(p_a)
    
    ## Experiment parameters
    N = 10000     # number of samples
    steps = 25    # the target time
    counter = 0  # to count the number of times the event {X_GameA  = 1} occurs

    ## Simulation
    for i in range(N):
        mc2.start()  # new experiment
        for j in range(steps):  mc2.move()
        if mc2.running_state == 'GameA':  counter += 1

    phat = counter / N
    y_axis.append(phat)

```

```python
plt.figure()
plt.plot(x_axis, y_axis)
# Specify some extra attributes
plt.xlabel('p')
plt.ylabel('X_GameA = 1')
plt.title('Probability of GameA happening')
```

### Ερμηνέια των αποτελεσμάτων

Αυτό συμβαίνει γιατί σύμφωνα με τις πιθανότητες μετάβασης ανώτερης τάξης, όταν η πιθανότητα να κερδίσει πόντο ο Α είναι μικρή, σε μεγαλύτερα n, η επίδραση της πολλαπλασιάζεται. Ομοίως και για μεγάλες τιμές της p. Όταν οι δύο τιμές είναι περίπου ίσες, η πιθανότητα νίκης του καθενός είναι περίπου ίδια.

## Παραδοτέο 2

```python
from simple_markov_chain_lib import markov_chain  # import markov chain simulator

# Transition Table
markov_table = {
    0: {1: .5, 2: .5},  # from state 0 we move to state 1 with prob 0.5 and to state 2 with 0.5
    1: {0: 1/3, 3: 2/3},
    2: {2: 1.},
    3: {0: .5, 3: .25, 4: .25},
    4: {4: 1.}
}

# Initial Distribution
init_dist = {0: 1.}  # we start from state 0 with probability 1

mc = markov_chain(markov_table, init_dist)
```

```python
M = 30
N = np.linspace(5, 12,500)
```

```python
variances = np.zeros(len(N))
sample_sizes = np.zeros(len(N))

for k in range(len(N)):
    sample_size = int(2 ** N[k])  # Ν
    sample_sizes[k] = sample_size
    
    values = np.zeros(M)
    for j in range(M):
        running_total = 0
        
        
        for i in range(sample_size):
            mc.start()
            while mc.running_state != 2 and mc.running_state != 4:
                mc.move()
            running_total += mc.steps  # steps it took to be absorbed

        mc_estimate = running_total / sample_size
        values[j] = mc_estimate
        # print("The estimated absorption time is %.2f steps" % mc_estimate)
    variance = stat.variance(values)
    variances[k] = variance
```

```python
plt.figure(figsize=(12, 7))  # define figure size

plt.subplot(1, 2, 1)  # setup subplots 1-row, 2-cols, select the 1st
plt.plot(sample_sizes, variances)
# Specify some extra attributes
plt.xlabel('Sample Size')
plt.ylabel('Variance of Estimator')
plt.title('Normal plot')
plt.grid(True)

plt.subplot(1, 2, 2) # select the 2nd subplot
plt.loglog(sample_sizes, variances, basex=2)

plt.xlabel('Sample Size')
plt.ylabel('Variance of Estimator')
plt.title('Logarithmic Plot')
plt.grid(True)  # add grid-lines
```
